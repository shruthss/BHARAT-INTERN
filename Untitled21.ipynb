```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a CSV file containing the Titanic dataset
titanic_data = pd.read_csv('/content/titanic dataset.csv')

# Create a bar plot for the count of survivors
survival_counts = titanic_data['Survived'].value_counts()
survival_counts.plot(kind='bar', color=['skyblue', 'salmon'])

# Customize the plot
plt.title('Survival Count on Titanic')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'], rotation=0)
plt.show()
```
